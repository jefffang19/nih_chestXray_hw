# do histagram equalization to all the images

```python
path = '../darknet/data/NIH/yolo/'

import os
img_txt = os.listdir(path)

imgs = []

for i in img_txt:
    if i[-4:] == '.png':
        imgs.append(i)
```

```python
imgs
```

```python
import cv2

for _path in imgs:
    src = cv2.imread(path + _path)

    if src is None:
        print('Could not open or find the image:')
        exit(0)

    src = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)
    dst = cv2.equalizeHist(src)
    
    cv2.imwrite(path + _path, dst)
```
